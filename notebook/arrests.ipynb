```python
import pandas as pd
```

```python
arr = pd.read_csv(r'C:\Users\river\Ironhack-data\Proyectos\ETL\data\NYPD_Arrest_Data__Year_to_Date_.csv')
```

```python
arr.head()
```

```python
arr = arr.drop(columns=['PD_CD', 'PD_DESC', 'KY_CD', 'LAW_CODE','ARREST_PRECINCT','JURISDICTION_CODE','PERP_SEX', 'PERP_RACE', 'X_COORD_CD', 'Y_COORD_CD', 'OFNS_DESC']
       )
```

```python
arr
```

```python
arr.rename(columns={
        'ARREST_KEY':'Arrest_id', 
        'ARREST_DATE':'Date',
        'OFNS_DESC':'Offence',
        'LAW_CAT_CD':'Offence',
        'ARREST_BORO':'Borough',
        'AGE_GROUP':'Age',
        

        }, inplace=True) #replace columns
```

```python
arr['Year'] = arr.Date.str[6:] #get first four values of Date.string
arr['Month'] = arr.Date.str[:2] #get the next 3 values of Date.string 
```

```python
arr
```

```python
mes_dic = {
    "01": "January",
    "02": "February",
    "03": "March",
    "04": "April",
    "05": "May",
    "06": "June",
    "07": "July",
    "08": "August",
    "09": "September",
    "10": "October",
    "11": "November",
    "12": "December"
}

arr['Month'] = arr['Month'].map(mes_dic)
```

```python
arr.Offence.unique()
```

```python
arr = arr.loc[(arr['Offence'] != '9') & (arr['Offence'].notnull())]
```

```python
arr.Borough.unique()
```

```python
borough_map = {'S': 'Staten Island',
               'M': 'Manhattan',
               'B': 'Brooklyn',
               'Q': 'Queens',
               'K': 'Bronx'}

# Sustituir los valores de la columna 'borough' por los nombres de los boroughs
arr['Borough'] = arr['Borough'].replace(borough_map)
```

```python
offence_map = {'F': 'Felony',
               'M': 'Misdemeanor',
               'V': 'Violation',
               'I': 'Infraction'}

# Sustituir los valores de la columna 'offence' por los tipos de Offence correspondientes
arr['Offence'] = arr['Offence'].replace(offence_map)
```

```python
arr
```
